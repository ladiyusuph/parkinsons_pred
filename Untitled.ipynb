## Data Preprocessing

```python
#Seperating the data into features and label
X = parkinsons_df.drop(columns = ['name','status'],axis = 1)
y = parkinsons_df.status
```

```python
#Splitting the data into training and test data
SEED = 90
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size = 0.2, random_state = SEED)

#Validating the split
print(X.shape, X_train.shape, X_test.shape)
print(y.shape, y_train.shape, y_test.shape)
```

```python
#Standardization of the data 
scaler = StandardScaler()

#Fitting the and transforming the data
X_train_scaled = scaler.fit_transform(X_train) 
X_test_scaled = scaler.transform(X_test)
```

## Model Building

```python
#Creating an instance of SVC
model = svm.SVC(kernel = 'linear')

#Training the model with training data
model.fit(X_train_scaled, y_train)
```

## Model Evaluation

```python
#Evaluating the model on the training data
X_train_pred = model.predict(X_train_scaled)

train_acc = accuracy_score(y_train, X_train_pred)

#Printing the model accuracy
print(f"The accuracy of the model on the training data is: {train_acc}")
```

```python
#Evaluating the model on the test data
X_test_pred = model.predict(X_test_scaled)

test_acc = accuracy_score(y_test, X_test_pred)

#Printing the model accuracy
print(f"The accuracy of the model on the training data is: {test_acc}")
```

## Building a Prediction System

```python
#Passing in a known row from our dataframe to see if the ML Model can rightly classify the the data
input_data = (119.99200,157.30200,74.99700,0.00784,0.00007,0.00370,0.00554,0.01109,0.04374,0.42600,0.02182,0.03130,0.02971,0.06545,0.02211,21.03300,0.414783,0.815285,-4.813031,0.266482,2.301442,0.284654)

#Converting the input data into a numpy array
input_array = np.asarray(input_data)

#Reshaping the array
input_reshaped = input_array.reshape(1,-1)

#Standardizing the input data
input_scaled = scaler.transform(input_reshaped)

#Making a prediction
input_prediction = model.predict(input_scaled)
print(input_prediction)

if (input_prediction[0] == 0):
    print('The Individual does not have Parkinsons Disease')
else:
    print('The Individual has Parkinsons Disease')
```

The Model was able to correctly label the input data
